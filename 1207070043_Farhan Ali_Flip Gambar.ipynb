Menngimport Library

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
# URL gambar yang akan diambil
url = "https://th.bing.com/th/id/OIP.FKCkiwltIzBLvicwzy-JjwHaEo?w=301&h=188&c=7&r=0&o=5&pid=1.7"
# Membaca gambar dari URL
img = imageio.imread(url)

# Menampilkan gambar
plt.imshow(img)
plt.show()
```

Mendapatkan resolusi

```python
# Mendapatkan informasi dimensi dan tipe gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
# Menampilkan informasi dari gambar
print("Tinggi Gambar :", img_height)
print("Lebar Gambar :", img_width)
print("Channel Gambar :", img_channel)
print("Tipe Gambar :", img_type)
```

Membuat variabel dengan resolusi dan tipe yang sama seperti gambar

```python
#np.zeros(img.shape, img_type) digunakan untuk membuat array kosong dengan dimensi yang sama dengan gambar img. 
#Fungsi np.zeros() digunakan untuk membuat array berisi nol dengan dimensi yang ditentukan.
img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)
```

Membalik gambar secara horizontal

```python
#Pada perulangan pertama (for y in range(0, img_height)), kita melakukan iterasi untuk setiap baris dalam gambar img. Variabel y akan berisi indeks baris saat ini.
for y in range(0, img_height):
#Pada perulangan kedua (for x in range(0, img_width)), kita melakukan iterasi untuk setiap kolom dalam gambar img. Variabel x akan berisi indeks kolom saat ini.
    for x in range(0, img_width):
#Pada perulangan ketiga (for c in range(0, img_channel)), kita melakukan iterasi untuk setiap saluran warna dalam gambar img. Variabel c akan berisi indeks saluran warna saat ini.
        for c in range(0, img_channel):
#melakukan pemutaran horizontal pada gambar img dan menyimpan hasilnya dalam variabel img_flip_horizontal
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]
```

Membalik gambar secara vertical

```python
# Iterasi untuk setiap baris dalam gambar
for y in range(0, img_height):
  # Iterasi untuk setiap kolom dalam gambar
    for x in range(0, img_width):
      # Iterasi untuk setiap saluran warna dalam gambar
        for c in range(0, img_channel):
          # Mengisi img_flip_vertical dengan piksel dari posisi berlawanan pada sumbu vertikal
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]
```

Menampilkan hasil balik gambar

```python
plt.imshow(img_flip_horizontal)  # Menampilkan gambar hasil pemutaran horizontal
plt.title("Flip Horizontal")  # Menetapkan judul plot sebagai "Flip Horizontal"
plt.show()  # Menampilkan plot gambar

plt.imshow(img_flip_vertical)  # Menampilkan gambar hasil pemutaran vertikal
plt.title("Flip Vertical")  # Menetapkan judul plot sebagai "Flip Vertical"
plt.show()  # Menampilkan plot gambar
```
