Import Library

```
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

Membaca Gambar

```
# Mengambil gambar dari  web
img = imageio.imread("https://th.bing.com/th/id/OIP.ti7pwt3r1q05xJGAYEHqtAHaEY?pid=ImgDet&rs=1")

# Menampilkan gambar
plt.imshow(img)
plt.show()
     
```

```
# Mengambil data dari gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

Merubah gambar menjadi Grayscale

```
img_grayscale = np.zeros(img.shape, dtype=np.uint8)  # Membuat array kosong dengan dimensi yang sama dengan gambar, dengan tipe data uint8

for y in range(0, img_height):  # Iterasi untuk setiap baris dalam gambar
    for x in range(0, img_width):  # Iterasi untuk setiap kolom dalam gambar
        red = img[y][x][0]  # Mendapatkan nilai komponen merah piksel
        green = img[y][x][1]  # Mendapatkan nilai komponen hijau piksel
        blue = img[y][x][2]  # Mendapatkan nilai komponen biru piksel
        gray = (int(red) + int(green) + int(blue)) / 3  # Menghitung rata-rata dari komponen merah, hijau, dan biru untuk mendapatkan nilai skala abu-abu
        img_grayscale[y][x] = (gray, gray, gray)  # Menetapkan nilai skala abu-abu yang sama ke setiap komponen piksel

plt.imshow(img_grayscale)  # Menampilkan gambar skala abu-abu
plt.title("Grayscale")  # Menetapkan judul plot sebagai "Grayscale"
plt.show()  # Menampilkan plot gambar
```

**Menampilkan Histogram Gambar Grayscale**

Membuat variabel untuk menyimpan data gambar

```
hg = np.zeros((256)) #hg dengan panjang 256 yang diinisialisasi dengan nilai nol. Array ini bertujuan untuk digunakan sebagai histogram grayscale.
```

Mengisi setiap nilai dalam array hg dengan 0

```
for x in range(0, 256):  # Iterasi untuk setiap nilai intensitas piksel dari 0 hingga 255
    hg[x] = 0  # Menginisialisasi setiap elemen dalam array hg dengan nilai nol
```

Menghitung nilai dari gambar

```
for y in range(0, img_height):  # Iterasi untuk setiap baris dalam gambar
    for x in range(0, img_width):  # Iterasi untuk setiap kolom dalam gambar
        gray = img_grayscale[y][x][0]  # Mendapatkan nilai intensitas piksel skala abu-abu
        hg[gray] += 1  # Menambahkan 1 pada elemen ke-gray dalam array hg
```

Menampilkan Histogram

```
plt.figure(figsize=(20, 6))
plt.plot(hg, color="black", linewidth=2.0)
plt.show()

bins = np.linspace(0, 256, 100)
plt.hist(hg, bins, color="black", alpha=0.5)
plt.title("Histogram")
plt.show()
```

Menghitung nilai dari gambar

```
for x in range(0, 256):  # Iterasi untuk setiap elemen dalam rentang 0-255
    hgr[x] = 0  # Mengatur nilai elemen ke-x dalam array hgr menjadi 0
    hgg[x] = 0  # Mengatur nilai elemen ke-x dalam array hgg menjadi 0
    hgb[x] = 0  # Mengatur nilai elemen ke-x dalam array hgb menjadi 0
    
for x in range(0, 768):  # Iterasi untuk setiap elemen dalam rentang 0-767
    hgrgb[x] = 0  # Mengatur nilai elemen ke-x dalam array hgrgb menjadi 0

temp = [0]  # Membuat array temp dengan elemen awal 0

for y in range(0, img.shape[0]):  # Iterasi untuk setiap baris dalam gambar
    for x in range(0, img.shape[1]):  # Iterasi untuk setiap kolom dalam gambar
        red = int(img[y][x][0])  # Mendapatkan nilai intensitas merah
        green = int(img[y][x][1])  # Mendapatkan nilai intensitas hijau
        blue = int(img[y][x][2])  # Mendapatkan nilai intensitas biru
        green = green + 256  # Menambahkan offset 256 ke intensitas hijau
        blue = blue + 512  # Menambahkan offset 512 ke intensitas biru
        hgrgb[red] += 1  # Meningkatkan frekuensi histogram untuk intensitas merah
        hgrgb[green] += 1  # Meningkatkan frekuensi histogram untuk intensitas hijau
        hgrgb[blue] += 1  # Meningkatkan frekuensi histogram untuk intensitas biru

binsrgb = np.linspace(0, 768, 100)  # Membuat array bins untuk plot histogram dengan rentang 0-768
plt.hist(hgrgb, binsrgb, color="black", alpha=0.5)  # Menampilkan histogram RGB
plt.title("Histogram Red Green Blue")
plt.show()
```

**Menampilkan Histogram Kumulatif**

```
hgk = np.zeros((256))  # Array untuk menyimpan frekuensi histogram grayscale
c = np.zeros((256))  # Array untuk menyimpan nilai kumulatif

for x in range(0, 256):  # Menginisialisasi array hgk dan c dengan nilai 0
    hgk[x] = 0
    c[x] = 0

for y in range(0, img_height):  # Iterasi untuk setiap baris dalam gambar
    for x in range(0, img_width):  # Iterasi untuk setiap kolom dalam gambar
        gray = img_grayscale[y][x][0]  # Mendapatkan nilai intensitas grayscale
        hgk[gray] += 1  # Meningkatkan frekuensi histogram grayscale

c[0] = hgk[0]  # Menginisialisasi nilai kumulatif dengan frekuensi histogram grayscale pada intensitas 0
for x in range(1, 256):  # Menghitung nilai kumulatif untuk setiap intensitas grayscale
     c[x] = c[x-1] + hgk[x]

hmaxk = c[255]  # Mencari nilai maksimum dalam nilai kumulatif

for x in range(0, 256):  # Normalisasi nilai kumulatif ke rentang 0-190
    c[x] = 190 * c[x] / hmaxk

plt.hist(c, bins, color="black", alpha=0.5)  # Menampilkan histogram kumulatif grayscale
plt.title("Histogram Grayscale Kumulatif")
plt.show()
```
